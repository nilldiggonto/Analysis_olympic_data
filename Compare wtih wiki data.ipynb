```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('medal_table.csv')
```

```python
data.head()
```

```python
data.set_index('index',inplace=True)
```

```python
data.head()
```

```python
data_new_grouped = data.groupby(['Year','Country','Medal'])
```

```python
data_new_unstack = data_new_grouped['Medal'].count().unstack(fill_value=0)
```

```python
medal_tables = data_new_unstack[['Gold','Silver','Bronze']]
```

```python
medal_tables
```

```python
data_1976 = pd.read_csv('data_1976.csv',index_col='Country')
```

```python
data_1976.head()
```

```python
##comparing with 1976
```

```python
agg_1976 = medal_tables.loc[1976].sort_values(['Gold','Silver','Bronze'],ascending=False).copy()
```

```python
agg_1976.head()
```

```python
div_76 = agg_1976.sub(data_1976).abs().dropna()
```

```python
div_76
```

```python
score_76 = div_76.sum()
```

```python
score_76
```

```python
score_76.sum()
```
